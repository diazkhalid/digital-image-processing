```python
import matplotlib.pyplot as plt #melakukan import library matplotlib.pyplot sebagai plt
import pandas as pd #melakukan import library pandas sebagai pd
import numpy as np #melakukan import library numpy sebagai np
import cv2 as cv #melakukan import library cv2 sebagai cv
```

Jurnal Nomor 1

```python
imgplt = plt.imread("benzema.jpg") #mengimport file benzema.jpg kedalam variabel imgplt
plt.title("RGB Image")
plt.imshow(imgplt) #menampilkan citra yang telah di import ke dalam variabel imgplt
```

```python
grayimgplt = cv.cvtColor(imgplt, cv.COLOR_BGR2GRAY) #mengonversi warna dari BGR(dalam format cv2) ke GRAY
plt.title("Grayscale Image")
plt.imshow(grayimgplt, cmap='gray') #menampilkan hasil citra yang telah di konversi 
```

```python
grayimgplt
```

```python
tinggi, lebar = grayimgplt.shape #shape pada numpy untuk melihat dimensi dari citra pada grayimgplt dan disimpan pada variabel tinggi dan lebar
print(tinggi, lebar) #mencetak variabel tinggi dan lebar untuk menampilkan dimensi dari citra pada grayimgplt
```

```python
sumGray = np.zeros(256).astype(int) #menggunakan zeros untuk membuat array 1 dimensi yang berisi 0 sebanyak 256
#melakukan perulangan untuk menghitung jumlah pixel yang memiliki derajat keabuan i
for  i in range(tinggi): 
    for j in range(lebar): 
        val = grayimgplt[i,j] 
        sumGray[val] += 1 
print ("ni:\n", sumGray) #mencetak jumlah pixel yang memiliki derajat keabuan i
print("n:", sumGray.sum()) #mencetak jumlah seluruh pixel pada citra
```

```python
grayValue = [i for i in range (256)] #melakukan perulangan untuk membuat array yang berisi angka 0 sampai 255 dan menyimpannya di variabel grayValue
print ("i:", grayValue) #mencetak grayValue atau derajat keabuan
```

```python
plt.ylabel("Jumlah Pixel") #memberikan label "Jumlah Pixel" pada sumbu y  
plt.xlabel("Derajat Keabuan") #memberikan label "Derajat Keabuan" pada sumbu x 
plt.bar(grayValue, sumGray) #mencetak histogram skala warne bardasarkan jumlah warna keabuan
```

```python
hi = sumGray/(tinggi*lebar) #melakukan perhitungan terhadap probabilitas piksel dan menyimpan hasilnya dalam variabel hi
print ("hi:\n", hi)
```

```python
plt.ylabel("Peluang Pixel") #memberikan label "Peluang Pixel" pada sumbu y
plt.xlabel("Derajat Keabuan") #memberikan label "Derajat Keabuan" pada sumbu x
plt.bar(grayValue, hi) #mencetak histogram skala warna berdasarkan probabilitas piksel
```

```python
print(grayimgplt) #mencetak matrix citra dari grayimgplt
grayimgplt.shape #menampilkan dimensi dari citra grayimgplt
```

```python
kernel = np.array([ #membuat array dengan numpy dan memasukkan nilainya ke dalam variabel kernel
    [0, -1, 0],
    [-1, 5, -1],
    [0, -1, 0],
    [-1, 5, -1]
])
print("Kernel: \n", kernel) #mencetak array yang telah disimpan pada variabel kernel
```

```python
padding = np.pad(grayimgplt, [[1, 1],[1, 1]], 'constant') #melakukan zero padding pada citra grayimgplt, yaitu menambahkan nilai 0 di sekeliling citra 
print ("Citra dengan padding\n", padding[0:len(grayimgplt), 0:len(grayimgplt[0])]) #mencetak citra grayimgplt setelah dilakukannya zero padding
```

```python
konvolusi = np.zeros([len(grayimgplt),len(grayimgplt[0])]) #membuat matrix 0 dengan zeros berukuran panjang dan lebar citra grayimgplt dan menyimpannya pada variabel konvolusi
konvolusiClip = np.zeros([len(grayimgplt),len(grayimgplt[0])]) #membuat matrix 0 dengan zeros berukuran panjang dan lebar citra grayimgplt dan menyimpannya pada variabel konvolusi
for i in range(1,len(grayimgplt)-1): # melakukan perulangan sebanyak panjang citra-1 (karena telah dilakukan padding sebelumnya)
    for j in range (1,len(grayimgplt[0])-1): # melakukan perulangan sebanyak lebar citra-1 (karena telah dilakukan padding sebelumnya)
        #konvolusi dilakukan dengan menambahkan hasil dari perkalian matriks citra setelah padding dengan kernel (sum of product)
        a = padding[i-1,j-1]*kernel [0,0] + padding[i-1,j]*kernel [0,1] + padding[i-1,j+1]*kernel [0,2] 
        b = padding[i,j-1]*kernel [1,0] + padding[i,j]*kernel [1,1] + padding[i,j+1]*kernel [1,2]
        c = padding[i+1,j-1]*kernel [2,0] + padding[i+1,j]*kernel [2,1] + padding[i+1,j+1]*kernel [2,2]
        konvolusi[i-1,j-1] = a+b+c
        konvolusiClip[i-1, j-1] = konvolusi[i-1, j-1]
        #melakukan clipping, yaitu dengan mengganti setiap nilai citra yang berada di bawah 0 manjadi 0 dan yang berada di atas 255 menjadi 255
        if (konvolusiClip[i-1, j-1] < 0):
            konvolusiClip [i-1, j-1] = 0
        elif (konvolusiClip[i-1, j-1] > 255):
            konvolusiClip [i-1, j-1] = 255 
```

```python
plt.title("Konvolusi Tanpa Clipping")
plt.imshow(konvolusi,cmap="gray") #menampilkan hasil citra grayimgplt setelah dilakukan konvolusi
```

```python
plt.title("Konvolusi Dengan Clipping")
plt.imshow(konvolusiClip,cmap="gray") #menampilkan hasil citra grayimgplt setelah dilakukan konvolusi dengan clipping
```

```python
x = 200
y = 250
trans_ = np.zeros([len(grayimgplt)+x, len(grayimgplt[0])+y])
for a in range(len(grayimgplt)):
    for b in range (len(grayimgplt[0])):
        trans_[a+x][b+y] = grayimgplt[a][b]
plt.imshow(trans_, cmap = "gray")
print(trans_)
```

```python
Tk = 200
Tl = 250
trans = np.zeros([len(grayimgplt), len(grayimgplt[0])])
for k in range (len(grayimgplt)):
    for l in range (len(grayimgplt[0])):
        new_k = k - Tk
        new_l = l - Tl
        if(new_k >= len(grayimgplt) or new_l >= len(grayimgplt[0]) or new_k < 0 or new_l < 0):
            trans[k][l] = 0
        else:
            trans[k][l] = grayimgplt[new_k][new_l]
plt.imshow(trans, cmap="gray") 
print (trans)
```
