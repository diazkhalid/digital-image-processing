```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rgb2gray(img):
    imgHeight = len(img)
    imgWidth = len(img[0])
    grayImg = np.empty([imgHeight, imgWidth], dtype=np.uint8)

    for i in range (imgHeight):
        for j in range (imgWidth):
            grayImg[i][j] = int((img[i][j][0]*0.2126+img[i][j][1]*0.7152+img[i][j][2]*0.0722)/3)
    
    return grayImg
```

```python
bangHaji = plt.imread("D:/File Kuliahan/Semester 4/PCD/Praktikum/Praktikum PCD/kelompok-1/Modul 1/benzema.jpg")
plt.imshow(bangHaji)
```

```python
grayBangHaji = rgb2gray(bangHaji)
```

```python
plt.imshow(grayBangHaji, cmap='gray')
```
